```python
# Read all files names in output folder and create a dictionary with the similar file names in a list


# First we will create a list with all the files in the output folder
# We will use the glob module to do this
# For each file we are going to use split each file name with the character "_"
# And the join again from the sixth character to the end of the string
# Then we will use a dictionary to save the list of files with the same name
# We will check if the file name is already in the dictionary
# If it is we will add the file name to the list
# If not we will create a new list with the file name

# Get all the files in the output folder

import glob

files = glob.glob("output/history/*")

# Create a dictionary to save the files with the same name

files_dict: dict[str, list[str]] = {}

# Loop through all the files in the output folder

for file in files:
    # Split the file name with the character "_"
    file_name = file.split("_")
    # Join the file name from the sixth character to the end of the string
    file_name = "_".join(file_name[6:])
    # Check if the file name is already in the dictionary
    if file_name not in files_dict:
        # If it is we will add the file name to the list
        files_dict[file_name] = [file]
        continue
    
    # If not we will create a new list with the file name
    files_dict[file_name].append(file)

print(files_dict)
```

```python
# Now we want to compare the files in the same list
# We will use the filecmp module to do this
# We will loop through the dictionary
# For each list of files we will compare the files
# If the files are the same we will delete the file
# If the files are different we will keep the file

from ast import Set
import filecmp
import hashlib
import os

# Loop through the dictionary

deleted_files = Set()

for file_list in files_dict.values():
    # Loop through the list of files
    for file in file_list:
        # Loop through the list of files again
        for file2 in file_list:
            # Compare the files
            if filecmp.cmp( file, file2, shallow=True):
                print("The files are the same")
                print(file)
                print(file2)
                print(hashlib.md5(open(file,'rb').read()).hexdigest())
                print(hashlib.md5(open(file2,'rb').read()).hexdigest())
                # If the files are the same we will delete the file
                # os.remove(file2)
                deleted_files.add(file2)
            continue
        continue
    continue
```
